```python
import calc as cl
```

```python
print("Digite qual o número correspondente à operação que deseja realizar:")

print("1 - Soma")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")
print("5 - Potenciação")
print("6 - Raiz")

op = int(input("Opção: "))

if op == 6:
    num = float(input("Digite o número: "))
    print(f"O resultado da raiz é: {cl.raiz(num)}")

else:
    a = float(input("Digite o primeiro número: "))
    b = float(input("Digite o segundo número: "))

    if op == 1:
        print("Resultado:", cl.soma(a, b))
    elif op == 2:
        print("Resultado:", cl.subtracao(a, b))
    elif op == 3:
        print("Resultado:", cl.multiplicacao(a, b))
    elif op == 4:
        print("Resultado:", cl.divisao(a, b))
    elif op == 5:
        print("Resultado:", cl.potencia(a, b))
    else:
        print("Opção inválida")
```
